Streaming data is notoriously nested. We employ specialized string manipulation to "explode" the `genres` and `production_countries` columns, which are often stored as lists.
````python
outlier_cols = ['runtime', 'imdb_votes', 'tmdb_popularity', 'imdb_score','tmdb_score']

for col in outlier_cols:
    Q1 = df_genres[col].quantile(0.25)
    Q3 = df_genres[col].quantile(0.75)
    IQR = Q3 - Q1

    upper_limit = Q3 + 1.5 * IQR
    lower_limit = Q1 - 1.5 * IQR


    upper_outliers = df_genres[df_genres[col] > upper_limit].shape[0]
    lower_outliers = df_genres[df_genres[col] < lower_limit].shape[0]

    print(f"--- {col} ---")
    print(f"Upper Limit: {upper_limit:.2f} | Outliers: {upper_outliers}")
    print(f"Lower Limit: {lower_limit:.2f} | Outliers: {lower_outliers}")
    print("-" * 20)
````

````python
def apply_final_capping(df, columns):
    # 1. Work on a fresh copy to break the link to the exploded view
    clean_df = df.copy()

    for col in columns:
        Q1 = clean_df[col].quantile(0.25)
        Q3 = clean_df[col].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # 2. Strategic Floor: Don't allow negative votes or popularity
        if col in ['imdb_votes', 'tmdb_popularity', 'runtime']:
            lower_bound = max(0, lower_bound)

        # 3. Permanent Imputation
        clean_df.loc[clean_df[col] < lower_bound, col] = lower_bound
        clean_df.loc[clean_df[col] > upper_bound, col] = upper_bound

        print(f"Fixed {col}: Range set to [{lower_bound:.2f}, {upper_bound:.2f}]")

    return clean_df

# Execute and re-assign
df_genres = apply_final_capping(df_genres, ['runtime', 'imdb_votes', 'tmdb_popularity', 'imdb_score','tmdb_score'])
````

### What all manipulations have you done and insights you found?

I. Structural Integrity & CleaningRelational Linking: Established a connection between titles.csv and credits.csv using the id field, allowing for future talent-based performance analysis.

Deduplication: Identified and removed 56 duplicate rows in the Credits dataset and 3 duplicate rows in the Titles dataset. This prevents "ghost data" from inflating counts in our Univariate analysis.

Handling Null Values: Logic-Based Imputation: Set seasons to 0 for all titles labeled as 'MOVIE', as nulls in this column were structurally expected rather than missing data.

* Statistical Imputation: Filled missing imdb_score, tmdb_popularity, and imdb_votes using Medians rather than means to protect the dataset from being skewed by extreme high/low values.

* Categorical Padding: Labeled missing age_certification and character names as 'Unknown' to maintain categorical completeness for plotting.

II. Feature Engineering & Explosion

- List Transformation: Used ast.literal_eval to convert string-represented lists (e.g., ['drama', 'action']) into actual Python lists.
- Granular Explosion: Performed a .explode() on the genres column.Impact: This transformed the dataset from a "per-title" grain to a "per-genre" grain, increasing the count from 9,868 to 22,269 rows. This ensures that a "Sci-Fi Thriller" is counted in both the Sci-Fi and Thriller categories for accurate genre-popularity metrics.

III. Advanced Outlier Management (Winsorization)
- IQR Detection: Calculated the Interquartile Range ($IQR = Q3 - Q1$) for numerical columns to define "normal" data boundaries.
- Strategic Capping: Instead of deleting outliers—which would remove Amazon's most successful "Blockbuster" hits—I implemented Capping (Winsorization).

1 Logic: Values above $Q3 + 1.5 \times IQR$ were capped at the upper bound, and values below $Q1 - 1.5 \times IQR$ were capped at the lower bound (with a floor of 0 for counts).

2 Reasoning: This creates a "level playing field" for visualizations. Without this, your charts would have a massive empty space between the average show and a 10-million-vote hit.
