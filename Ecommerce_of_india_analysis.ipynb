```python
# Define mapping for Age ranges
age_map = {
    "Less than 20 years": 18,
    "21-30 years": 25,
    "31-40 years": 35,
    "41-50 yaers": 45,   # note typo "yaers" in your data
    "51 years and above": 55
}

# Apply mapping
data["Age"] = data["Age"].map(age_map)

# Check results
print(data["Age"].unique())
```

```python
duration_map = {
    "Less than 1 year": 0.5,
    "1-2 years": 1.5,
    "2-3 years": 2.5,
    "3-4 years": 3.5,
    "Above 4 years": 5   # or 4.5 if you prefer midpoint
}

data["ShoppingDuration"] = data["ShoppingDuration"].map(duration_map)
```

```python
# Define mapping for PurchaseFrequency ranges
freq_map = {
    "Less than 10 times": 9.5,       # midpoint just below 10
    "11-20 times": 15.5,             # midpoint
    "21-30 times": 25.5,
    "31-40 times": 35.5,
    "41 times and above": 41.5,      # just above 41
    "42 times and above": 42.5       # just above 42
}

# Apply mapping
data["PurchaseFrequency"] = data["PurchaseFrequency"].map(freq_map)

# Check results
print(data[["PurchaseFrequency"]].head())
```

```python
data["Age"] = pd.to_numeric(data["Age"], errors="coerce")
```
